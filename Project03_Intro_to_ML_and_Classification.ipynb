# Intro

About the data analyzed:

Each line has a different website design, such as a school website, a bakery website, and so on. People interested in creating websites for their respective businesses submit the project to a company and stipulate the number of hours for the project to be created (expected_hours) , as well as the amount to be paid (price).

```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv"
dados = pd.read_csv(uri)

# renaming the data

a_renomear = {
    'expected_hours' : 'horas_esperadas',
    'price' : 'preco',
    'unfinished' : 'nao_finalizado'
}
dados = dados.rename(columns = a_renomear)
# manipulating the first column
trocar = {
    0 : 1,
    1 : 0
}

dados['finalizado'] = dados.nao_finalizado.map(trocar)

```

```python
import seaborn as sns 

sns.scatterplot(x= "horas_esperadas", y ="preco", data = dados)

```

```python
sns.scatterplot(x= "horas_esperadas", y ="preco", hue= "finalizado", data = dados)
```

```python
sns.relplot(x= "horas_esperadas", y ="preco",col data = dados)
```
